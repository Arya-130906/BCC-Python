```python
msg = input("Enter message: ")
count = 0
for ch in msg:
    if not ch.isalnum():
        count += 1
print(count)
```

```python
#finding total distance between the adjacent terms of a list of 5 elements
N=int(input()) #5
sum=0
myList=[] #[10,11,7,12,14]
for i in range(N):
  a=int(input("Enter the element: "))
  myList.append(a)
for j in range(len(myList)): #j=1
  if j+1 in range(len(myList)):
    sum+=abs(myList[j]-myList[j+1]) # 11-7=4
print("Total distance:",sum)
```
