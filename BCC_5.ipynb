```python
#how to return multiple values
def add():
  n1=int(input("enter number 1:"))
  n2=int(input("enter number 2:"))
  sum=n1+n2
  mul=n1*n2
  sub=n1-n2
  div=n1/n2
  return sum,mul,sub,div
result=add()
print(result)
```

```python
#Types od arguments
# positional argument
# keyword argument
# default argument
# variable length argument/variable number od argument
```

```python
#Positional Argument
def personalInfo(fname,lname):
    print("First Name=",fname)
    print("Last Name=",lname)
personalInfo("Arya","Paunikar")
```

```python
#Default argument
def cityName(city="Nagpur"):
  print(city)
cityName("Pune")
cityName("Mumbai")
cityName()
```

```python
#Variable Length argument
def studentName(*name):
  print(name)
  pass
studentName("Prashant","Rahul","sandip","ashish")
```

```python
#search the element 10
myList=[5,2,9,7,5,6]
def searchElement(target):
  for i in range(len(myList)):
    if target==myList[i]:
      return i
  return -1
result=searchElement(10)
if result !=-1:
  print("Element found at index number:",result)
else:
  print("Element not found.")
```
