```python
#WAP to accept three paper marks and check maximum marks using nested if else
a=int(input("Enter the first paper marks:"))
b=int(input("Enter the second paper marks:"))
c=int(input("Enter the third paper marks:"))
if a>b:
  if a>c:
    print("a is greatest")
  else:
    print("c is greatest")
else:
    if b>c:
      print("b is greatest")
    else:
        print("c is greatest")
```

```python
#counting vowels and consonents
name="arya"
data=['a','e','i','o','u']
vowels=0
con=0
for i in name:
  if i in data:
    vowels+=1
  else:
    con+=1
    print("vowels",vowels)
    print("consonent",con)
```
