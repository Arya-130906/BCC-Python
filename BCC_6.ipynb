```python
#Product of Array Except Self
nums = [1, 2, 3, 4]
result = []
for i in range(len(nums)):
    product = 1
    for j in range(len(nums)):
        if i != j:   # skip same index
            product = product * nums[j]
    result.append(product)
print(result)
```
